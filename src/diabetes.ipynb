```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Part A
df = pd.read_csv('../data_clean/diabetes_cleaned.csv')
np.random.seed(1094)
sample_data = df.sample(n=25)

mean_sample_glucose = sample_data['Glucose'].mean()
max_sample_glucose = sample_data['Glucose'].max()

total_data_glucose_mean = df['Glucose'].mean()
total_data_glucose_max = df['Glucose'].max()

plt.bar(['25 Records Sample Max', 'Total Records Max'], [mean_sample_glucose, total_data_glucose_mean])
plt.title('Glucose Means')
plt.ylabel('Glucose')
plt.show()

plt.bar(['25 Records Sample Max', 'Total Records Max'], [max_sample_glucose, total_data_glucose_max])
plt.title('Glucose Max Values')
plt.ylabel('Glucose')
plt.show()
```

```python
#Part B
sample_data_bmi_98th_percentile = np.percentile(sample_data['BMI'], 98)
total_data_bmi_98th_percentile = np.percentile(df['BMI'], 98)

plt.bar(['Sample 98th Percentile BMI', 'Total Population 98th Percentile BMI'], 
        [sample_data_bmi_98th_percentile, total_data_bmi_98th_percentile])
plt.title('98th Percentile of BMIs')
plt.ylabel('BMI')
plt.show()
```

```python
#Part C
num_of_samples = 500
each_sample_size = 150

bootstrap = np.zeros((num_of_samples, 3))

for i in range(num_of_samples):
    sample_data= df['BloodPressure'].sample(n=each_sample_size, replace=True)
    bootstrap[i, 0] = sample_data.mean()
    bootstrap[i, 1] = sample_data.std()
    bootstrap[i, 2] = np.percentile(sample_data, 90)

total_data_bp_mean = df['BloodPressure'].mean()
total_data_bp_std = df['BloodPressure'].std()
total_data_bp_percentile = np.percentile(df['BloodPressure'], 90)

plt.bar(['Bootstrap Mean', 'Total Records Mean'], [bootstrap[:, 0].mean(), total_data_bp_mean])
plt.title('BloodPressure Means')
plt.ylabel('BloodPressure')
plt.show()

plt.bar(['Bootstrap Standard Deviation', '90th Percentile of Total Given Records'], 
        [bootstrap[:, 1].mean(), total_data_bp_std])
plt.title('BloodPressure Standard Deviations')
plt.ylabel('BloodPressure')
plt.show()

plt.bar(['Bootstrap Data 90th Percentile', '90th Percentile of Total Given Records'], 
        [bootstrap[:, 2].mean(), total_data_bp_percentile])
plt.title('BloodPressure Data 90th Percentiles')
plt.ylabel('BloodPressure')
plt.show()
```
